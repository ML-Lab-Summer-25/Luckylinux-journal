# Lab: Model Representation
### This notebook demonstrates a simple linear model using NumPy and Matplotlib to understand how training data and model parameters interact.

## 1. Importing Libraries
We start by importing NumPy and setting the Matplotlib style.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

## 2. Training Data
We define simple training examples using NumPy arrays.

```python
#Training data in the form of 1D numpy array
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
```

```python
m = x_train.shape[0]
print(f"Number of training Data : {m}")
```

```python
#Plotting training data using Matplotlib
plt.scatter(x_train, y_train, c='r', marker='o')

plt.title('Real Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')

plt.show()
```

## 3. Model Parameters and Visualization
We visualize different linear models using matplotlib.

```python
#These are the parameters that has to be figured out with the help of training data
w = 200
b = 100
```

```python
#Function to plot the graph based on the above parameters and features(x_train) from the training data
def model_output(x, w, b):
    y = np.zeros(x.shape[0])

    for i in range(x.shape[0]):
        y[i] = (w*x[i]) + b

    return y
```

```python
temp_y = model_output(x_train, w, b)

plt.plot(x_train, temp_y, label='Model prediction', c='b')
plt.scatter(x_train, y_train, c='r', marker='o', label='Actual values')

plt.title("Housing Prices")

plt.ylabel('Price (in 1000s of dollars)')

plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
#Now this is the house size for which our model has to predict price
x_i = 1.9 #in 1000 sqft.
y_i = (x_i*w + b)
print(f"Value of {x_i*1000}sqft. house is : ${y_i*1000}")
```

```python
#Final plot to mark the training data, model predicted graph, Predicted price for x_i
plt.plot(x_train, temp_y, label='Model prediction', c='b')
plt.scatter(x_train, y_train, c='r', marker='o', label='Actual values')
plt.scatter(x_i, y_i, label='Predicted price', c='g', marker='X')

plt.title("Housing Prices")

plt.ylabel('Price (in 1000s of dollars)')

plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
